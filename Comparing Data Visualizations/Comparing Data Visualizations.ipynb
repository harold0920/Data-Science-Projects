```python
# Import the needed libraries
import pandas as pd
import matplotlib.pyplot as plt

# Load the dataset

df = pd.read_csv("countries.csv")
df.head()
```

```python
# Filter the dataset for the year 1952 only
df_1952 = df[df["year"] == 1952]

# Filter life expectancy data for Asia
asia_life = df_1952[df_1952["continent"] == "Asia"]["lifeExpectancy"]

# Filter life expectancy data for Africa
africa_life = df_1952[df_1952["continent"] == "Africa"]["lifeExpectancy"]

# Display the first few values from each to verify
print("Asia Life Expectancy (1952):")
print(asia_life.head())

print("\nAfrica Life Expectancy (1952):")
print(africa_life.head())
```

```python
# Create histograms using subplot() to compare Asia and Africa
plt.figure(figsize=(12, 5))

# Histogram for Asia
plt.subplot(1, 2, 1)
plt.hist(asia_life, bins=10)
plt.title("Asia Life Expectancy (1952)")
plt.xlabel("Life Expectancy")
plt.ylabel("Frequency")

# Histogram for Africa
plt.subplot(1, 2, 2)
plt.hist(africa_life, bins=10)
plt.title("Africa Life Expectancy (1952)")
plt.xlabel("Life Expectancy")
plt.ylabel("Frequency")

plt.tight_layout()
plt.show()
```
