```python
# Import libraries needed for analysis
import seaborn as sns
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Load the diamonds dataset from seaborn
diamonds = sns.load_dataset("diamonds")

# Display the first few rows
diamonds.head()
```

```python
# Check dataset information
diamonds.info()

# Summary statistics
diamonds.describe()
```

```python
# Univariate Analysis
# I chose the variable "price" to analyze the distribution of diamond prices.

# Histogram of price
plt.figure(figsize=(8,5))
sns.histplot(diamonds["price"], bins=30, kde=True)

plt.title("Distribution of Diamond Prices")
plt.xlabel("Price")
plt.ylabel("Frequency")
plt.show()
```

```python
# Boxplot to visualize spread and outliers in price

plt.figure(figsize=(6,4))
sns.boxplot(x=diamonds["price"])

plt.title("Boxplot of Diamond Prices")
plt.show()
```

```python
# Multivariate Analysis
# I chose the variables "carat" and "price" to see the relationship between the size of the diamond and its price.

plt.figure(figsize=(8,5))
sns.scatterplot(x="carat", y="price", data=diamonds)

plt.title("Relationship Between Carat and Price")
plt.xlabel("Carat")
plt.ylabel("Price")
plt.show()
```

```python
# Correlation analysis between numeric variables

plt.figure(figsize=(8,6))
sns.heatmap(diamonds.corr(numeric_only=True), annot=True, cmap="coolwarm")

plt.title("Correlation Between Diamond Features")
plt.show()
```

## Analysis:

1. Diamond prices are highly skewed with many lower priced diamonds.
2. There is a strong positive relationship between carat and price.
3. Larger diamonds tend to have significantly higher prices.
