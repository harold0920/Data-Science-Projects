```python
import pandas as pd
```

```python

df = pd.read_csv("Realestate.csv")

#Normalization
# Min-Max Normalization formula:
# (value - min) / (max - min)

df["houseAgeStandardized"] = (
    (df["houseAge"] - df["houseAge"].min()) /
    (df["houseAge"].max() - df["houseAge"].min())
)
```

```python


# 4. Drop the numberOfConvenientStores column and apply the change permanently to the dataset

df.drop(columns=["numberOfConvenienceStores"], inplace=True)

# 5. Change the transaction column name to transactionDate

df.rename(columns={"transaction": "transactionDate"}, inplace=True)


# 6. Use .loc[] to display all rows from 0 to 10

print("Rows 0 to 10 using .loc[]")
print(df.loc[0:10])


# 7. Use .iloc[] to display the first 10 rows

print("\nFirst 10 rows using .iloc[]")
print(df.iloc[0:10])


# 8. Find and remove duplicate rows

# Check how many duplicate rows exist and remove it
print("\nNumber of duplicate rows before removal:", df.duplicated().sum())
df.drop_duplicates(inplace=True)

# Check duplicates again
print("Number of duplicate rows after removal:", df.duplicated().sum())


# 7. Find missing values and fill them with the mean

# Display missing values per column
print("\nMissing values before filling:")
print(df.isnull().sum())

# Fill missing numeric values with column mean
df.fillna(df.mean(numeric_only=True), inplace=True)

# Display missing values after filling
print("\nMissing values after filling:")
print(df.isnull().sum())


# 8. Save the cleaned dataset so all modifications persist
df.to_csv("Real_Estate_Cleaned.csv", index=False)

print("\nData cleaning complete. Cleaned file saved as 'Real_Estate_Cleaned.csv'")
```
